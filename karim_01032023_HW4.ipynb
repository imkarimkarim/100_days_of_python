```python
import pytz
from datetime import datetime

name = input('- tell me you\'r name my Cali friend\n+ ')

utc_time = datetime.now(pytz.utc)
america_losAngeles = utc_time.astimezone(pytz.timezone('America/Los_Angeles'))

h = america_losAngeles.hour
m = america_losAngeles.minute
month = america_losAngeles.month
day = america_losAngeles.day


def greet(time_of_day):
    print('- Good', time_of_day, name)


if h >= 6 and h <= 11 and m <= 59:
    greet('morning')
elif h >= 12 and h < 16:
    greet('afternoon')
elif h >= 16 and h < 20:
    greet('evening')
elif h >= 20 and h < 23:
    greet('night')
else:
    print('- Sweet dreams', name)

if month == 3 and day == 1:
    print('- Happy birthday', name)
```
